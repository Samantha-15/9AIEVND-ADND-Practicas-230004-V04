```python
12.3 Funciones que mandan llamar otras funciones y que regresaron el resultado de un procesamiento, en este caso utilizaremos una funcion predefinida
en python denominada replace que nos permite sustitur letras en un cadena de texto
```

```python
def sustituir_vocales(texto):
    texto = texto.replace("A", "4")
    texto = texto.replace("a", "4")

    texto = texto.replace("E", "3")
    texto = texto.replace("e", "3")

    texto = texto.replace("I", "4")
    texto = texto.replace("i", "4")

    texto = texto.replace("O", "o")
    texto = texto.replace("o", "o")

    texto = texto.replace("U", "∪")
    texto = texto.replace("u", "∪")

    return texto
```

```python
print(sustituir_vocales(saludar_nombre("Sammy")))
```

**18. Eliminar los daros de los cuatiles menores**
Con la finalidad de hacer un analisis centrado en los datos tendencia es decir mayor frecuencia eliminamos los registros que

En este paso trabajamos con la columna MILES, que contiene datos numéricos (cuantitativos).
El objetivo es eliminar valores muy pequeños o muy grandes que podrían afectar el análisis. Lo cuales permiten realizar 
operaciones

```python
a=df['MILES'].quantile(0.95)
b=df['MILES'].quantile(0.05)
df3=df[df['MILES']> b]
df3=df[df['MILES']<a]
df3
```

**20. Estimar el vaor de Z-score**
Con la finalidad de estimar la disprecion de datos, es decir que tan variados o diferentes son, utilizamos esta formula estadistica
para poder continuar con el analisis EDA(Exploratory data analisis)

```python
df3['Z']=(df.MILES-df.MILES.mean())/df.MILES.std()
df3
```

**25. Grafica lugar de abordaje**
